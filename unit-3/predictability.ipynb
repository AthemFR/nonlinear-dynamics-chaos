## Predictability in chaotic systems

Suppose we are trying to predict the future state of a chaotic system to within a tolerance of $a = 10^{-3}$. Given that our estimate of the initial state is uncertain to within $\|\delta_0\| = 10^{-7}$, for about how long can we predict the state of the system, while remaining within the tolerance?

Now suppose we buy the finest instrumentation, recruit the best graduate students, etc., and somehow manage to measure the initial state a million times better, i.e., we improve our initial error to $\|\delta_0\| = 10^{-13}$. How much longer can we predict?


We may  use: 

$$ t_{\text{horizon}} \sim O\!\left( \frac{1}{\lambda} \ln \frac{a}{\|\delta_0\|} \right) $$



```python
# Importing libraries
import numpy as np
```

```python
# Horizon of predictability

a   = 1.e-3
d_0 = 1.e-7
d_00 = 1.e-13

t_h0 = np.log(a/d_0)

t_h00 = np.log(a/d_00)

print(t_h0, t_h00)
```

```python
print(f"After a {round(d_0/d_00, 4)} improvement in our initial uncertainty, we can ONLY predict {t_h00/t_h0} times longer.")
```
