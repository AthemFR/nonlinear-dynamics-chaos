# 3D Dynamical Systems:

Look at the references and choose **one** of the following **3D dynamical systems**:

## **Rössler System**

$$\dot{x} = -y - z$$  
$$\dot{y} = x + a y$$  
$$\dot{z} = b + z(x - c)$$

#### Reference: https://link.springer.com/chapter/10.1007/978-3-031-95228-9_19

## **Chen System**

$$\dot{x} = a(y - x)$$  
$$\dot{y} = (c - a)x - x z + c y$$  
$$\dot{z} = xy - b z$$

#### Reference: https://www.researchgate.net/publication/242919493_Yet_Another_Chaotic_Attractor

## **Sprott System**

$$\dot{x} = b y + c z$$  
$$\dot{y} = -x + a y$$  
$$\dot{z} = x^2 - z$$

#### Reference: https://wippl.wisc.edu/pub_files/journal/Sprott112.pdf
