### Dynamical systems:

In this assignment, choose **one** of the following **2D dynamical systems**:

- **Lotka–Volterra Predator–Prey Model**  

$$\dot{x} = x(\alpha - \beta y)$$ 

$$\dot{y} = -y(\gamma - \delta x)$$

- **Nonlinear Pendulum (Damped or Undamped)**  

$$\dot{\theta} = \omega$$
  
$$\dot{\omega} = -\sin\theta - q\,\omega$$

- **Rayleigh Oscillator (Planar Form)**

$$\dot{x} = y$$

$$\dot{y} = \mu(1 - x^2)y - x$$

- **Glycolytic Oscillator (Sel'kov Model)**  
$$\dot{x} = -x + a y + x^2 y$$

$$\dot{y} = b - a y - x^2 y$$

- **Brusselator Model**  

$$\dot{x} = a - (b+1)x + x^2 y$$
$$\dot{y} = bx - x^2 y$$
