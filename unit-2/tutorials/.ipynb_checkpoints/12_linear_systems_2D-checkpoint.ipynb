# Linear Systems

## Example: Simple harmonic oscillator

The vibrations of a mass hanging from a spring are governed by the linear ODE:

$$m\ddot{x}+kx=0$$

where $m$ is the mass, $k$ is the spring constant, and $x$ is the displacement of the mass.

(a) Use SymPy to solve this ODE.

(b) Provide a phase plane analysis of this simple harmonic oscillator.

#### (a) Use SymPy to solve this ODE.

```python
# Import libraries
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define symbols
t, m, k = sp.symbols('t m k', real = True, positive = True )

# Define the function x(t)
x = sp.Function('x')

# ODE:
diff_eq1 = sp.Eq(m * sp.Derivative(x(t), t, 2) + k * x(t), 0)

display(diff_eq1)
```

```python
# Solution
sln = sp.dsolve(diff_eq1 , x(t))

display(sln)
```

```python
# Lambdify the sln
rhs = sln.rhs

# Define symbols for C1 and C2
C1, C2 = sp.symbols('C1 C2')

# Fix the parameters: take fiducial values
sub_dict = {m: 1., k: 2., C1: 0.8, C2: 2.}

# Evaluate the RHS 
rhs_subs = rhs.subs(sub_dict)

# Lambdification 
x_func = sp.lambdify(t, rhs_subs, "numpy")

#print(type(x_func))
```

```python
# Generate vectors
t_axis = np.linspace(0., 3*np.pi, 1000)

# Evaluate our lamdified function
x_axis = x_func(t_axis)
```

```python
# Plotting
plt.figure(figsize =(11, 3))

plt.plot(t_axis, x_axis)

plt.xlabel(r"$t$")
plt.ylabel(r"$x(t)$")

plt.show()
```

#### (b) Provide a phase plane analysis of this simple harmonic oscillator.

##### Original ODE:

$m\ddot{x}+kx=0$

##### System of ODEs:

$\dot{x}=v$

$\dot{v}=-\frac{k}{m}x=-\omega^2x$


Every pair: $(x,v)$ -> $(\dot{x}, \dot{v})= (v, -\frac{k}{m}x)$

```python
# We want to represent v vs. x
# Create vectors
x = np.linspace(-20., +20., 20)
v = np.linspace(-20., +20., 20)

# Meshgrid
x_2d, v_2d = np.meshgrid(x, v)

# Fix constants
m_1 = sub_dict[m]
k_1 = sub_dict[k]
```

$\dot{x}=v$

$\dot{v}=-\frac{k}{m}x=-\omega^2x$

```python
# Function for the derivatives

def vector_field(x, v):
    """
    """
    x_dot= v
    v_dot = -k_1*x/m_1
    
    return x_dot, v_dot
```

```python
# Call the function 
x_dot2d, v_dot2d = vector_field(x_2d, v_2d)
```

```python
# Modulus of the vector field
vel = np.sqrt(x_2d**2 +  v_2d**2)
```

```python
# Phase Plot

plt.figure(figsize=(4,4))

plt.quiver(x_2d, v_2d, x_dot2d, v_dot2d, vel)

plt.plot(0,0, marker = "o", color = "blue")

plt.xlabel(r"$x$")
plt.ylabel(r"$v$")
plt.grid()

plt.show()
```

```python
# Phase Plot
st_val = np.linspace(1., 12., 10)

st_points = np.array([[s, 0] for s in st_val])

plt.figure(figsize=(4,4))

plt.streamplot(x_2d, v_2d, x_dot2d, v_dot2d, start_points=st_points, color = vel, density = 2.0)

plt.plot(0,0, marker = "o", color = "blue")

plt.xlabel(r"$x$")
plt.ylabel(r"$v$")
plt.grid()

plt.show()
```
