# Linear Systems

## Example: Simple harmonic oscillator

The vibrations of a mass hanging from a spring are governed by the linear ODE:

$$m\ddot{x}+kx=0$$

where $m$ is the mass, $k$ is the spring constant, and $x$ is the displacement of the mass.

(a) Use SymPy to solve this ODE.

(b) Provide a phase plane analysis of this simple harmonic oscillator.

#### (a) Use SymPy to solve this ODE.

```python
import sympy as sp 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Define symbols
t, m, k = sp.symbols('t m k', real=True, positive=True)

# 2. Define the function x as a function of t
x = sp.Function('x')

# 3. Define the ODE: m*d^2x/dt^2 + kx = 0
# Note: Derivative(x(t), t, 2) represents the second derivative, d^2x/dt^2
ode = sp.Eq(m * sp.Derivative(x(t), t, 2) + k * x(t), 0)

# 4. Solve the ODE
solution = sp.dsolve(ode, x(t))

# 5. Print the solution
display(solution)

# 6. Lambdify
rhs = solution.rhs # Get RHS
C1, C2 = sp.symbols('C1 C2') # Define Symbols
sub_dict = {m: 0.1, k: 6., C1: 0.5, C2: 2.} # Constants
rhs_subs= rhs.subs(sub_dict) # Evaluate using constants
x_function = sp.lambdify(t, rhs_subs, "numpy") # Lambdify

#print(type(x_function))
```

```python
# 7. Generate data for plotting
t_vals = np.linspace(0, 4 * np.pi, 2000)
x_vals = x_function(t_vals)
```

```python
# 6. Plot the result
plt.figure(figsize=(11, 3))

plt.plot(t_vals, x_vals, label=f'$x(t)$ for $m={m_val}$, $k={k_val}$')

plt.xlabel(r'$t$')
plt.ylabel(r'$x(t)$')

plt.show()
```

#### (b) Provide a phase plane analysis of this simple harmonic oscillator.

##### Original ODE:

$m\ddot{x}+kx=0$

##### System of ODEs:

$\dot{x}=v$

$\dot{v}=-\frac{k}{m}x=-\omega^2x$


```python
# Vectors
x = np.linspace(-10., 10., 10)
v = np.linspace(-10., 10., 10)

# Meshgrid
x_2d, v_2d = np.meshgrid(x, v)

# Fix constants
k = 2.
m = 1.

# Function
def dot_func(x,v):
    """
    """
    x_dot = v
    v_dot = -k*x/m
    
    return x_dot, v_dot
```

```python
# Function call

x_dot2d, v_dot2d = dot_func(x_2d, v_2d)

print(x_dot2d.shape)
```

```python
speed = np.sqrt(x_dot2d**2 + v_dot2d**2)
```

```python
plt.figure(figsize=(6,5))

plt.quiver(x_2d, v_2d, x_dot2d, v_dot2d, speed)# color = "red")

plt.plot(0,0, marker = "o", color = "blue")

plt.grid()

plt.show()
```

```python
s_values = np.linspace(1.0, 7.0, 10) # 10 amplitudes from A=1.0 to A=9.0
start_points = np.array([[s, 0] for s in s_values])

plt.figure(figsize=(6,5))

plt.streamplot(x_2d, v_2d, x_dot2d, v_dot2d, start_points=start_points, color = speed, density=8.0, cmap="viridis", zorder=1)

plt.plot(0,0, marker = "o")

plt.grid()

plt.xlim(-10, 10)
plt.ylim(-10, 10)


plt.show()
```
